## Transformando las variables explicativas

Los algoritmos de aprendizaje automático no comprenden datos en formato de texto, por lo que debemos transformar los datos a un formato numérico para que el algoritmo pueda interpretar la información. Esta transformación debe realizarse de manera que no altere la información original del conjunto de datos, por lo que no basta con simplemente cambiar los valores a números aleatorios.

image.png

````python
from sklearn.compose import make_column_transformer  # Este modulo ayuda a realizar una transformación de columnas
from sklearn.preprocessing import OneHotEncoder # Ayuda a realizar transformación de 1 y 0
````

````python
columnas = X.columns # Creamos una variable llamada columnas
````

````python
#Esto crea un ColumnTransformer (atajo con make_column_transformer) que one-hot-encoda las columnas ['estado_civil','escolaridad','default','prestatario'] y deja el resto de columnas sin transformar (passthrough). El resultado final se fuerza a una matriz densa (no sparse) y se desactiva una comprobación/advertencia relacionada con el formato de las columnas de remainder.
# make_column_transformer(...)Qué es: es un shorthand (atajo) para construir un ColumnTransformer. Permite aplicar distintos transformadores a subconjuntos de columnas y concatenar los resultados.
# (OneHotEncoder(drop='if_binary'), ['estado_civil', ...])  OneHotEncoder: transforma variables categóricas en variables dummy (una columna por categoría). Por defecto OneHotEncoder produce una matriz sparse salvo que se le indique lo contrario.
# drop='if_binary': le dice al encoder que solo para variables con exactamente 2 categorías elimine una de las columnas (es decir, de dos categorías genera 1 dummy en vez de 2). Esto evita colinearidad perfecta en casos binarios (útil para regresiones lineales sin regularización). Para variables con más de 2 categorías no se elimina nada.
# Lista de columnas: ['estado_civil','escolaridad','default','prestatario'] son las columnas sobre las que se aplicará ese encoder.
#Nota práctica: si existe la posibilidad de categorías nuevas en datos de transform (no vistas en fit), considera handle_unknown='ignore' para evitar errores en producción.
# remainder='passthrough' Indica que las columnas no listadas en los transformadores deben pasarse tal cual (sin transformar) y concatenarse con la salida de los transformadores especificados. Si no pones esto, por defecto las otras columnas se eliminan (drop).
#sparse_threshold=0:
#Controla si la salida final del ColumnTransformer será sparse (matriz dispersa) o densa (numpy array).
#Regla general: si hay mezcla de salidas sparse/densas, scikit-learn calculará la densidad y si esa densidad es menor que sparse_threshold devolverá matriz sparse; de lo contrario devolverá densa.
#sparse_threshold=0 fuerza que siempre se devuelva una matriz densa (útil si quieres evitar trabajar con matrices scipy.sparse).
#Scikit-learn
#force_int_remainder_cols=False
#Parámetro introducido en versiones recientes para controlar el formato interno/alarmas cuando se usa remainder='passthrough'.
#Ponerlo en False evita ciertas advertencias sobre el formato de las columnas del remainder en transformers_.
#(Ha sido añadido y su comportamiento/valor por defecto ha ido cambiando y está en proceso de deprecación en versiones más nuevas). En la práctica: si tienes warnings al usar remainder='passthrough', fijar force_int_remainder_cols=False suele silenciarlos.
one_hot = make_column_transformer((OneHotEncoder(drop='if_binary'),
                                  ['estado_civil','escolaridad','default','prestatario']),
                                  remainder='passthrough',
                                  sparse_threshold=0,
                                  force_int_remainder_cols=False)
#sparse_threshold=0 fuerza que siempre se devuelva una matriz densa (útil si quieres evitar trabajar con matrices scipy.sparse).
#Scikit-learn
#force_int_remainder_cols=False
#Parámetro introducido en versiones recientes para controlar el formato interno/alarmas cuando se usa remainder='passthrough'.
#Ponerlo en False evita ciertas advertencias sobre el formato de las columnas del remainder en transformers_.
#(Ha sido añadido y su comportamiento/valor por defecto ha ido cambiando y está en proceso de deprecación en versiones más nuevas). En la práctica: si tienes warnings al usar remainder='passthrough', fijar force_int_remainder_cols=False suele silenciarlos.
````

````python
X = one_hot.fit_transform(X)

# one_hot.fit(X) y one_hot.transform(X) (o fit_transform)normalmente devuelven un ndarray (denso) o scipy.sparse según la configuración (sparse_threshold, OneHotEncoder.sparse_output, etc.). En tu caso sparse_threshold=0 forzará un ndarray denso.
````

````python
one_hot.get_feature_names_out(columnas)

#sirve para obtener los nombres de las columnas que genera tu transformador (one_hot) después de aplicarlo a las columnas originales que pasaste (en este caso, columnas).
````

````python
X
````

````python
pd.DataFrame(X, columns=one_hot.get_feature_names_out(columnas))
````

# 4. Selección de modelos

## Normalizando los datos

Algunos algoritmos pueden asignar un mayor peso a los valores de las variables debido a la escala de los valores y no por la importancia de la clasificación de la variable objetivo. Por ejemplo, en una base de datos con las columnas **edad** y **salario**, el algoritmo puede dar un mayor peso de decisión a los valores del salario simplemente por estar en una escala mayor que los valores de la edad, y no porque la variable salario sea más importante que la variable edad.

En estos casos, necesitamos realizar una transformación en los datos para que estén en una misma escala, evitando que el algoritmo sea influenciado incorrectamente por los valores numéricos divergentes entre las variables.

$X_{sc} = \frac{X - X_{min}}{X_{max} - X_{min}} $

## KNN

El algoritmo KNN se basa en el cálculo de la distancia entre los registros de la base de datos y busca elementos que estén cerca unos de otros (vecinos) para tomar la decisión de clasificación.

Debido a que utiliza cálculos de distancia, este algoritmo está influenciado por la escala de las variables, y por eso es necesario realizar una transformación en los datos antes de utilizar este método.
